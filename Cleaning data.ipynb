## Create a new dataframe with two columns: date and value

```python
#put all values in one column + create the date column as: d = date(year, month, day)
values=[]
cal=[]
for i in range(1,13):
    for j in range(682):
        condition= level_00_21.iloc[:,i+1].isnull().values
        if condition[j] == False:
            #print(level_00_21.iloc[j,i+1])
            values.append(level_00_21.iloc[j,i+1])
            
            day=level_00_21.iloc[j,1]
            y=level_00_21.iloc[j,0]
            d = date(y, i, day)
            cal.append(d)
        

```

```python
#print(len(cal), len(values))

rhein_level = pd.DataFrame({'date': cal, 'level': values})
display(rhein_level)
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(x=rhein_level['date'], y=rhein_level['level'])
```
